```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fmri=sns.load_dataset("fmri")
fmri.head(5)
```

```python
sns.relplot(data=fmri,x='timepoint', y='signal')
```

```python
sns.relplot(x='timepoint', y='signal',data=fmri,kind="line",sort=True)
```

```python
sns.catplot(x='timepoint', y='signal', kind='box', data=fmri)
plt.show()
```
